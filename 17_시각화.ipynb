```python
dates = ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05']
prices = [100, 102, 101, 105, 103]
```

```python
plt.figure(figsize=(10,6)) # 차트 사이즈 설정
plt.title('5일간 주식가격')
plt.ylim(95,110)
plt.xlabel('Date')
plt.ylabel('Prices')
plt.grid()

plt.plot(dates, prices, label='Stock Price')
plt.legend()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(dates, prices, color='green', marker='o') # line
plt.xticks(rotation=40)

plt.subplot(1,2,2)
plt.bar(dates, prices, color='orange') # bar
plt.xticks(rotation=40)

plt.savefig('my_stock_price_chart.png')
```
