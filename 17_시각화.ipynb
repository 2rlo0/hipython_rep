# 새로운 데이터셋
seaborn.tips 데이터셋 - 식사 가격, 팁 가격, 사람의 성별, 흡연석?, 요일, 시간, 테이블 사이즈(인원)

```python
import seaborn as sns
tips_df = sns.load_dataset('tips')
tips_df.head(10)
```

```python
# 1변량 그래프: 히스토그램
```

```python
plt.hist(data=tips_df, x='total_bill', bins=20)
```

```python
fig = plt.figure(figsize=(12,5))
axes1= fig.add_subplot(1,2,1) # total_bill
axes2=fig.add_subplot(1,2,2) # tip
axes1.hist(data=tips_df, x='total_bill', bins=20)
axes2.hist(data=tips_df, x='tip', bins=20)
plt.tight_layout()
plt.show()
```

```python
# total_bill으로 10~20달러로 많이 지불함. 많은 비용을 지불하는 식당은 아님.
# tip으로는 2~4달러로 많이 지불함. 총금액에 상관없이 팁을 2달러로 많이 지불함.
```

```python
# boxplot
fig = plt.figure(figsize=(12,5))
axes= fig.add_subplot(1,1,1) # total_bill
axes.boxplot(data=tips_df, x='total_bill')
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes= fig.add_subplot(1,1,1) # total_bill
axes.boxplot(
    x=[
        tips_df[tips_df['sex']=='Female']['total_bill'],
        tips_df[tips_df['sex']=='Male']['total_bill']
    ],
    labels=['Female', 'Male']
)
plt.show()
# 총금액값이 남성이 여성보다 최대값도 높고 평균값도 높음
```

```python
fig = plt.figure(figsize=(12,5))
axes= fig.add_subplot(1,1,1)
axes.boxplot(
    x=[
        tips_df[tips_df['sex']=='Female']['tip'],
        tips_df[tips_df['sex']=='Male']['tip']
    ],
    labels=['Female', 'Male']
)
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes= fig.add_subplot(1,1,1) 
axes.boxplot(
    x=[
        tips_df[tips_df['size']==1]['tip'],
        tips_df[tips_df['size']==2]['tip'],
        tips_df[tips_df['size']==3]['tip'],
        tips_df[tips_df['size']==4]['tip'],
        tips_df[tips_df['size']==5]['tip']
        
    ],
    labels=['1 size', '2 size', '3 size', '4 size', '5 size']
    
)
plt.title('인원별 팁')
plt.show()
# 인원수가 많을수록 팁의 분산도 커지는 경향
```

```python
fig = plt.figure(figsize=(12,5))
axes= fig.add_subplot(1,1,1) 
axes.boxplot(
    x=[
        tips_df[tips_df['time']=='Lunch']['tip'],
        tips_df[tips_df['time']=='Dinner']['tip']
    ],
    labels=['Lunch', 'Dinner'],
    
)
plt.show()
```

```python
fig = plt.figure(figsize=(12,5))
axes= fig.add_subplot(1,1,1) 
axes.boxplot(
    x=[
        tips_df[tips_df['day']=='Thur']['tip'],
        tips_df[tips_df['day']=='Fri']['tip'],
        tips_df[tips_df['day']=='Sat']['tip'],
        tips_df[tips_df['day']=='Sun']['tip']
    ],
    labels=['Thur', 'Fri', 'Sat', 'Sun'],
    
)
plt.show()
```

```python
tips_df.groupby(['day', 'time']).size()
```

```python
# scatter, 산점도

fig = plt.figure()
axes= fig.add_subplot(1,1,1)
axes.scatter(
    tips_df['total_bill'], tips_df['tip']
)
plt.show()
```

```python
colors = {"Female":"red", "Male":"blue"}
# 다변량 표현
# 성별에 따라 - 색상
# 값의 크기에 따라 - 점의 크기

fig = plt.figure()
axes= fig.add_subplot(1,1,1)
axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s=tips_df['size']**2*10, # tablesize, tip
    c=tips_df['sex'].map(colors)
)
plt.show()
```

```python
colors = {"Female":	"#DDA0DD", "Male":"#87CEEB"}
fig = plt.figure()
axes= fig.add_subplot(1,1,1)
axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s=tips_df['tip']**2*10, # tablesize, tip
    c=tips_df['sex'].map(colors),
    alpha=0.5
)
plt.show()
```

```python
tips_df.time.unique()
```

```python
colors = {"Female":"#FFB6C1", "Male":"#20B2AA"}
tips_df['tip_pct'] = tips_df['tip']/tips_df['total_bill']


fig = plt.figure()
axes= fig.add_subplot(1,1,1)
axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s=tips_df['tip_pct']*1000, # 크기값
    c=tips_df['sex'].map(colors) # 색상값
)
plt.show()
```

```python
# 파이 차트: 비율
# 금액, 사이즈 등


# 성별 비율 비교
# 빈도수
sex_counts  = tips_df['sex'].value_counts()
sex_counts.head()
```

```python
fig = plt.figure()
axes= fig.add_subplot(1,1,1)
axes.pie(
    sex_counts,
    colors=['#20B2AA', '#FFB6C1'],
    labels=sex_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.show()
```

```python
day_counts  = tips_df['day'].value_counts()
day_counts.head()
```

```python
fig = plt.figure()
axes= fig.add_subplot(1,1,1)
axes.pie(
    day_counts,
    colors=['#20B2AA', '#FFB6C1', '#DDA0DD','#87CEEB'],
    labels=day_counts.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.show()
```
