```python
import matplotlib.pyplot as plt

# ✅ 한글 폰트 설정 (Windows 사용자: Malgun Gothic, macOS: AppleGothic, Linux: NanumGothic 등)
plt.rcParams['font.family'] = 'Malgun Gothic'  # 또는 'AppleGothic', 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

# ✅ 데이터
regions = ['서울', '경기/인천', '충청도', '강원도', '경상도', '전라도', '제주도']
counts = [3893, 2924, 1188, 396, 3174, 1134, 130]
colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#edc948', '#b07aa1']

# ✅ 막대그래프 생성
plt.figure(figsize=(10, 6))
bars = plt.bar(regions, counts, color=colors)

# ✅ 제목 및 축 라벨
plt.title('지역별 공연건수', fontsize=16)
plt.xlabel('지역', fontsize=12)
plt.ylabel('공연건수', fontsize=12)

# ✅ 각 바 위에 수치 표기
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 100,
             f'{height:,}', ha='center', va='bottom', fontsize=11)

# ✅ 기타 설정
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

regions = ['서울', '경기/인천', '충청도', '강원도', '경상도', '전라도', '제주도']
counts = [3893, 2924, 1188, 396, 3174, 1134, 130]
colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#edc948', '#b07aa1']

plt.figure(figsize=(10, 6))
bars = plt.bar(regions, counts, color=colors)

# ✅ 제목 및 축 레이블 (글자 크게, 굵게)
plt.title('지역별 공연건수', fontsize=20, fontweight='bold')
plt.xlabel('지역', fontsize=14)
plt.ylabel('공연건수', fontsize=14)

# ✅ 축 글꼴 키우기
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# ✅ 수치 레이블 폰트 및 위치 조정
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height + 150,  # 100 → 150 으로 여유 확보
             f'{height:,}', ha='center', va='bottom', fontsize=13, fontweight='bold')

plt.tight_layout()
plt.show()
```
