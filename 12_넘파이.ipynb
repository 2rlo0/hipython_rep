```python
# 비교 연산 예제
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([1, 3, 3, 0, 5])

# 같음 비교
print("같음:", arr1 == arr2)

# 크기 비교
print("더 큼:", arr1 > arr2)
print("더 작음:", arr1 < arr2)

# 논리 연산 결합
mask = (arr1 >= 3) & (arr2 <= 4)
print("복합 조건:", mask)

# 비교 결과를 이용한 필터링
filtered = arr1[arr1 > 3]
print("3보다 큰 값:", filtered)
```

```python
# 일반 Python 반복문 vs NumPy 벡터화 연산 비교
import numpy as np
import time

# 큰 배열 생성
size = 1000000
arr1 = np.array(range(size))
arr2 = np.array(range(size))
result_list = [0] * size

# Python 반복문 방식
start_time = time.time()
for i in range(size):
    result_list[i] = arr1[i] + arr2[i]
python_time = time.time() - start_time

# NumPy 벡터화 연산 방식
start_time = time.time()
result_numpy = arr1 + arr2
numpy_time = time.time() - start_time

print(f"Python 반복문 실행 시간: {python_time:.4f}초")
print(f"NumPy 벡터화 연산 실행 시간: {numpy_time:.4f}초")
print(f"속도 향상: {python_time/numpy_time:.1f}배")
```

## 불리언 인덱싱
비교연산자, 논리연산자를 이용한 마스크 행렬을 생성

마스크 행렬을 이용한 인덱싱


```python
# 부울행렬
x = np.array([1,2,3,4,5])
x < 3  # 조건식
```

```python
np.less(x, 3) # 조건 함수
```

```python
x1 = np.array([[1,2,3],[4,5,6]])
x1.shape
x1 > 5
```

```python
data = np.arange(28).reshape(7,4)
data>5
```

```python
data[data>5] # 조건에 맞는 값 추출 > 1차원
```

```python
name = np.array(['Bob', 'Joe', 'Will', 'Bob', 'Will', 'Joe', 'Joe'])
name
```

```python
name =='Bob'
```

```python
data[name=='Bob'] # 부울 인덱싱
```

## 조건식의 규칙

- 조건식으로 사용할 행렬의 값이 문자여도 True/False 식으로 나오면 필터가 된다.
- 필터로 사용할 행렬의 갯수 == 원본 데이터의 갯수 : "짝이 맞아야한다" Indexerr
- 행렬[조건식, ]
- 행렬[:, 조건식] 앞에 생략할 수 없음
- 괄호() 로 우선순위를 지정할 수 있다.
- 복합조건: 논리함수 and, or > & | 로 표현한다.
- 부정연산자: != , ~

```python
name
name1 = np.array(['Bob','Joe', 'Will'])
data[name1 == 'Bob']  #2d array index [r,c]
```

```python
data[name=='Bob',]
```

```python
data[name=='Bob', :]
```

```python
data[name=='Bob',1]
```

```python
data[,name=='Bob'] # 뒤는 생략 가능, 앞은 x
```

```python
data, name
```

```python
mask = (name == 'Bob') | (name =='Will')
```

```python
data[mask]
```

```python
mask1 = (name != 'Bob') & (name !='Will')
```

```python
data[mask1]
```

```python
mask2 = ~(name == 'Joe')
data[mask2]
```

```python
mask3 = ~((name == 'Bob') | (name =='Will'))
data[mask3]
```
