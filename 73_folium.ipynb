```python
import requests
import folium
from pandas import json_normalize
```

```python
targetSite='https://www.starbucks.co.kr/store/getStore.do?r=8K3XF06R9Q'
req=requests.post(targetSite, data={
'ins_lat': 37.563398,
'ins_lng': 126.9863309,
'p_sido_cd': '01',
'p_gugun_cd': '',
'in_biz_cd': '',
'iend': 600,
'set_date': ''
})
starbucks=req.json()
print(type(starbucks))
```

```python
st_df = json_normalize(starbucks, 'list')
st_df
```

```python
st_df_map = st_df[['s_name','sido_code','sido_name','gugun_code','gugun_name',
'doro_address','lat','lot']]
st_df_map.info()
```

```python
st_df_map['lat'] = st_df_map['lat'].astype(float)
st_df_map['lot'] = st_df_map['lot'].astype(float)
st_df_map.info()
```

```python
st_df_map.head(5)
```

```python
loc1 = [37.501087, 127.043069]
stMap = folium.Map(location=loc1, zoom_start=18)
pop = folium.Popup('역삼아레나빌딩', max_width=100)
folium.Marker(location=loc1, popup=pop).add_to(stMap)
stMap
```

```python
# 위도, 경도만 뽑아서 보기 
st_df_lat_lot = st_df[['lat','lot']]
print(st_df_lat_lot)
#  [lat, lot] 한 쌍으로 리스트에 저장 
loc_list = st_df_lat_lot.values.tolist()
print(loc_list)
# s_name을 리스트에 순서대로 저장 
s_name_list = st_df_map['s_name'].tolist()
s_name_list
# for문 사용해서 전체 스벅 지도에 찍기 

# 지도 시작 위치 (첫 번째 매장 기준)
stMap = folium.Map(location=loc_list[0], zoom_start=13)

# 매장 이름(s_name_list)과 좌표(loc_list) 묶어서 for문
for i in range(len(s_name_list)):
    name = s_name_list[i]
    loc = loc_list[i]
    folium.Marker(location=loc, popup=name).add_to(stMap)
    
stMap
```
